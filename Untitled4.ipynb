```python
a = float(input())
b = float(input())
f = str(input())
if b == 0 and (f == "/" or f == "mod" or f == "div"):
  print ("Деление на 0!")
else:
  operations = {
    '+': lambda x, y: x + y,
    '-': lambda x, y: x - y,
    '*': lambda x, y: x * y,
    '/': lambda x, y: x / y,
    'mod': lambda x, y: x % y,
    'pow': lambda x, y: x ** y,
    'div': lambda x, y: x // y,
  }
  print (operations[f](a, b)) 
```

```python
a = int(input())
if a >=0 and a <=1000:
  if a >= 11 and a <= 14 or a >= 111 and a <= 114:
    print (str(a) + " программистов")
  elif  a == 1 or a % 10 == 1:
    print (str(a) + " программист")  
  elif a > 1 and a < 5 or a % 10 > 1 and a % 10 < 5:
    print (str(a) + " программиста")
  else:
    print (str(a) + " программистов")
```

```python
f = str(input())
if f == "треугольник":
  a = int(input())
  b = int(input())
  c = int(input())
  p = (a+b+c)/2
  print ((p*(p-a)*(p-b)*(p-c))**0.5)
elif f == "прямоугольник":
  a = int(input())
  b = int(input())
  print (a*b)
elif f == "круг":
  r = int(input())
  print (3.14*r*r)
else:
  print("Ошибка ввода")  
```
